```python
from pytadbit.parsers.hic_parser import load_hic_data_from_reads
import matplotlib.pyplot as plt
import numpy as np
reso=10000
```

```python
hic_data = load_hic_data_from_reads(reads,  reso)
matrix = hic_data.get_matrix(focus=None, diagonal=True, normalized=False)
```

```python
len(matrix[0])
```

```python
beg1, end1, beg2, end2 = map(lambda x: x / reso, [500000, 550000, 2000000, 2500000])
```

```python
nrows = len(matrix)
new_matrix = []
for k, m in enumerate(matrix):
    line = m[k:] + m[0:k]
    new_matrix.append(list(np.array(line[0:nrows/2]) + np.array(line[-nrows/2+1:][::-1])))
```

```python
plt.figure(figsize=(12, 400))
for k in xrange(0, nrows):
    plt.subplot(nrows, 1, k + 1)
#     plt.plot(np.log2(new_matrix[k]), alpha=1)
    plt.plot(new_matrix[k], alpha=1)
    pos1 = min(abs(beg2 - k), min(k, beg2) + (nrows - max(k, beg2)))
    pos2 = min(abs(end2 - k), min(k, end2) + (nrows - max(k, end2)))
    pos1, pos2 = sorted([pos1, pos2])
    if beg2 < k < end2:
        plt.fill_between([0, pos1], 0, 25000, alpha=0.3, color='green')
        plt.fill_between([0, pos2], 0, 25000, alpha=0.3, color='green')
    else:
        plt.fill_between([pos1, pos2], 0, 25000, alpha=0.3, color='green')
    pos3 = min(abs(beg1 - k), min(k, beg1) + (nrows - max(k, beg1)))
    plt.fill_between([pos3, pos3 + 1], 0, 25000, alpha=0.3, color='red')
    plt.ylim(0, 20000)
    plt.xlim(0, nrows / 2)
plt.show()
```
